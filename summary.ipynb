```python
import pandas as pd
import numpy as np
```

```python
data2=pd.read_csv('dataset.csv')
```

```python
data2.head()
```

```python
summ=data2['summary']
```

```python
summ
```

```python
from textblob import TextBlob
```

```python
def sentiment_calc(text):
    try:
        return TextBlob(text).sentiment.polarity
    except:
        return None
```

```python
data2['sentiment'] = summ.apply(sentiment_calc)
```

```python
data2.head()
```

```python
dataset2=data2[['summary','sentiment']]
```

```python
dataset2
```

```python
def polarity(x):
    if x >0:
        return 0
    return -1
y_sum = dataset2['sentiment']
#print(actualScore)
pol =y_sum.map(polarity)
#print(New_score)
y_sum=pol
```

```python
y_sum
```

```python
from sklearn.model_selection  import train_test_split
X_train, X_test, y_train, y_test = train_test_split(summ, y_sum, test_size = 0.25, random_state = 110)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
vectorizer = vect.fit(X_train)
```

```python
vect.vocabulary_
```

```python
X_train_transformed = vect.transform(X_train)
X_test_transformed =vect.transform(X_test)
```

```python
print(type(X_train_transformed))
print(X_train_transformed)
```

```python
from sklearn.svm import SVC
#svcclassifier = SVC()
svcclassifier = SVC(kernel = 'linear', random_state = 10,verbose= True)
```

```python
svcclassifier.fit(X_train_transformed, y_train)
```

```python
y_pred = svcclassifier.predict(X_test_transformed)
print(y_pred)
```

```python
print(y_pred)
print(y_test)
```

```python
from sklearn import metrics
confusion = metrics.confusion_matrix(y_test, y_pred)
print(confusion)
#[row, column]
TP = confusion[0, 0]
FN = confusion[0, 1]
FP = confusion[1, 0]
TN = confusion[1, 1]
```

```python
sensitivity = TP / float(FN + TP)
print("sensitivity",sensitivity)
```

```python
specificity = TN / float(TN + FP)

print("specificity",specificity)
```

```python
precision = TP / float(TP + FP)
print("Precision",precision)
```

```python
accuracy=(TP + TN) / (TP + TN + FP + FN)
print("Accuracy",accuracy)
```
